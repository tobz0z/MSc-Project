```python
def crop_black(path):
    img_test = tf.io.read_file(path)
    # img_test = tf.image.decode_jpeg(img_test, channels=3)
    img_test = tf.image.decode_png(img_test, channels=3)
    img_test = K.eval(img_test)
    img_test = img_test.squeeze()
    hang = []
    lie = []
    for inde1, j in enumerate(img_test):
        for inde2, k in enumerate(j):
            if k not in [0, 1, 2, 3]:
                hang.append(inde1)
                lie.append(inde2)

    hang_min = min(hang)
    hang_max = max(hang)
    lie_min = min(lie)
    lie_max = max(lie)

    img = cv2.imread(path)
    img1 = cv2.imread('./Colab Notebooks/data/train/images/' + path[36:])
    crop = img[hang_min:hang_max, lie_min:lie_max, :]
    crop1 = img1[hang_min:hang_max, lie_min:lie_max, :]
    cv2.imwrite('./Colab Notebooks/data2/train_del_black/images/' + path[36:], crop)
    cv2.imwrite('./Colab Notebooks/data2/train_del_black/labels/' + path[36:], crop1)
```

```python
def read_path():
    train_images = glob.glob('./Colab Notebooks/data/train/images/*.png')
    train_images.sort()
    return train_images
```

```python
image_path = read_path()
for i, j in enumerate(image_path):
    crop_black(j)
    print(i)
```
